### 4.1 Adding New column "profit".

Profit= ***NET SALES VALUE(t) - NET SALES VALUE(t-1) / NET SALES VALUE(t-1)***

```python
#Adding New Column "Profit"
sales_data['Profit'] = sales_data['Net Sales Value'].pct_change(periods=1)#pct_change is amethod to caluclate percentage.
sales_data
```

### 4.2 Conversion of Categorical Data into Dummy Variables

The most of the input variables or features are not showing direct linear relationship with outputvariable or feature, so we have to do some transformation to make linear.

#### Data transformation techniques if the data is continous.
1. Log Transformation.
2. Standard scalar.
3. Min-Max scalar.
4. Square root transformation and etc....

#### Data transformation techniques if the data is discrete.
1. Label-encoding.
2. One-Hot encoding

***One hot encoding*** : is a technique that we use to represent categorical variables as numerical values in a machine learning model. The advantages of using one hot encoding include: It allows the use of categorical variables in models that require numerical input.

***Label Encoding*** : makes it possible for algorithms to interpret and comprehend the data by giving each category within a variable a distinct number.

```python
sales_data.head()
```

```python
sales_data.dtypes
```

```python
# Transforming all categorical features into numerical data by using label encoder transformation method.
encoder = LabelEncoder()
sales_data['Day'] = encoder.fit_transform(sales_data['Day'])
sales_data['SKU'] = encoder.fit_transform(sales_data['SKU'])
sales_data['BU'] = encoder.fit_transform(sales_data['BU'])
sales_data['Brand'] = encoder.fit_transform(sales_data['Brand'])
sales_data['Model'] = encoder.fit_transform(sales_data['Model'])
sales_data
```

### 4.3 Standardization of Numerical Variables

#### 1. Why we need to do standardize?.

=> Because data standardization helps to improve the quality of our data by transforming and standardizing.

#### 2. When we have to do standardize?

=> When our data varying that's the time we should do standardize the data.

#### 3. How do we do standardize the data?

=> Using **Z-score**, is one of the most technique for to do standardize the data. And can be done by subtracting the mean and dividing by the standard deviation for each value of each of each feature.

```python
X = sales_data.drop(labels= 'Date',axis =1)
```

```python
X
```

```python
#Standardization using Z-scalar
scaled = StandardScaler()
scaled_data = scaled.fit_transform(X)
final_data = pd.DataFrame(data= scaled_data,columns=X.columns)
final_data
```
